```python
import pickle
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph data from the .pkl file
with open('knowledge_graph1.pkl', 'rb') as f:
    graph_data = pickle.load(f)

# Create a NetworkX graph
G = nx.Graph()

# Add nodes and edges based on graph_data structure
if isinstance(graph_data, list):  # Assuming graph_data is a list of tuples (node1, node2)
    G.add_edges_from(graph_data)
elif isinstance(graph_data, dict):  # Assuming graph_data is a dictionary (node1: [neighbors])
    for node, neighbors in graph_data.items():
        G.add_node(node)
        for neighbor in neighbors:
            G.add_edge(node, neighbor)

# Draw the graph using a spring layout for better visualization
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True) 

# Display the plot within the Jupyter Notebook
plt.show()
```

```python
import pickle
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline 
 # For displaying plots within the notebook

# Create a sample graph data 
sample_graph_data = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('B', 'D'),
    ('C', 'D')
]

# Save the sample data to a .pkl file
with open('knowledge_graph1.pkl', 'wb') as f:
    pickle.dump(sample_graph_data, f)

# Load the graph data from the .pkl file
with open('knowledge_graph1.pkl', 'rb') as f:
    graph_data = pickle.load(f)

# Create a NetworkX graph
G = nx.Graph()
G.add_edges_from(graph_data)

# Draw the graph using a spring layout for better visualization
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True) 

# Display the plot within the Jupyter Notebook
plt.show()
```
